```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
dataframe = pd.read_csv('activity.csv', parse_dates=['created_at'])
dataframe
```

```python
dataframe.dtypes
```

```python
dataframe['created_at_formatted'] = dataframe['created_at'].dt.strftime('%Y-%m')
```

```python
dataframe
```

```python
dataframe_count = dataframe.groupby(['created_at_formatted', 'device']).count()
dataframe_count
```

```python
x = dataframe_count.index.map(lambda x: x[0])
x = x.unique()
list(x)
```

```python
dataframe_count.dtypes
```

```python
grupo = dataframe_count.reset_index()
grupo
```

```python
y_android = grupo.loc[(grupo['device'] == 'android')]['id'].values
y_ios = grupo.loc[(grupo['device'] == 'ios')]['id'].values
y_mac = grupo.loc[(grupo['device'] == 'mac')]['id'].values
y_windows = grupo.loc[(grupo['device'] == 'windows')]['id'].values
```

```python
fig = plt.figure(figsize=(30,7))
ax = fig.add_subplot(111)

ax.stackplot(x, y_android, y_ios, y_mac, y_windows, edgecolor='white', labels=['android', 'ios', 'mac', 'windows'])

ax.legend(loc='upper left')
ax.set_title('Número de logins em dispositivos por mês')
ax.margins(0, 0) # Set margins to avoid "whitespace"

plt.show()
```

```python
fig = plt.figure(figsize=(30,7))
ax = fig.add_subplot(111)

x = list(set(grupo.created_at_formatted))

ax.bar(x, y_android, edgecolor='white', width=0.85, label="android")
ax.bar(x, y_ios, bottom=y_android, edgecolor='white', width=0.85, label="ios")
ax.bar(x, y_mac, bottom=[i+j for i,j in zip(y_android, y_ios)], edgecolor='white', width=0.85, label="mac")
ax.bar(x, y_windows, bottom=[i+j for i,j in zip(y_ios, y_mac)], edgecolor='white', width=0.85, label="windows")
 
ax.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)
ax.set_title('Número de logins em dispositivos por mês')
 
plt.show()
```
