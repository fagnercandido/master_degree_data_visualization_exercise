```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd
from datetime import date
```

```python
#https://api.covid19api.com/total/dayone/country/portugal/status/confirmed
```

```python
dataframe = pd.read_json('https://api.covid19api.com/total/dayone/country/portugal/status/confirmed')
```

```python
dataframe
```

```python
dataframe['Date']
```

```python
dataframe
```

```python
plt.figure(figsize=(100,20))

cases = dataframe['Cases'].diff().fillna(dataframe['Cases'])

plt.fill_between(dataframe['Date'], cases)
plt.plot_date(dataframe['Date'], cases,'-')

plt.rc('font',**{'family' : 'normal', 'weight' : 'bold', 'size' : 40})
plt.show()
```
